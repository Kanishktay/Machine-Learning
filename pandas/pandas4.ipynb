```python
def quality(total_bill,tip):
    if tip/total_bill>0.15:
        return 'Generous'
    else:
        return 'other'
```

```python
quality(16,4)
```

```python
# slow method
```

```python
df['tip_info'] = df[ ['total_bill','tip'] ].apply( lambda df : quality( df['total_bill'], df['tip']),axis = 1 )
```

```python
#fast method atleast 10 times
```

```python
df['tip_info'] = np.vectorize(quality)(df['total_bill'],df['tip'])
```
