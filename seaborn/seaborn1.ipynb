```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head(10)
```

```python
plt.figure(figsize=(12,4),dpi = 500)
sns.scatterplot(x = 'salary' , y = 'sales',data = df,size = 'work experience',alpha = 0.5)
```

```python
plt.figure(figsize=(12,4),dpi = 500)
sns.scatterplot(x = 'salary' , y = 'sales',data = df,style = 'level of education' ,hue ='level of education' ,s =50,alpha = 0.4)
plt.savefig('new.png')
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize = (7,4))
sns.rugplot(x = 'salary',data = df,height = 0.5)
```

```python
sns.set(style = 'ticks')
sns.displot(data = df,x = 'salary',bins = 50,color = 'red',edgecolor = 'blue',linewidth = 1,kde=True,rug=True)
```

```python
sns.histplot(data = df,x='salary')
```

```python
sns.kdeplot(data = df, x = 'salary')
```
