```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.jointplot(data = df, x = 'math score',y = 'reading score',hue = 'gender')
```

```python
sns.jointplot(data = df, x = 'math score',y = 'reading score',kind = 'kde',shade =True)
```

```python
sns.pairplot(data = df,hue = 'race/ethnicity', corner = True)
```

```python
sns.catplot(data = df,x= 'gender',y= 'reading score',kind = 'violin',col = 'lunch' , row = 'test preparation course')
```

```python
sns.pairplot(data =df)
```
