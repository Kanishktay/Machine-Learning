```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize = (10,4))
sns.boxplot(data =df,y='math score',x = 'parental level of education',hue = 'test preparation course')
plt.legend(loc = (1.05,0.5))
```

```python
plt.figure(figsize = (10,10))
sns.violinplot(data =df,x='reading score',y = 'parental level of education',hue = 'test preparation course',split = True,inner = 'quartile')
plt.legend(loc = (1.55,0.5))
```

```python
plt.figure(figsize = (10,10))
sns.violinplot(data =df,x='reading score',y = 'parental level of education',hue = 'test preparation course',
               split = True,inner = 'quartile',bw = 0.3)
plt.legend(loc = (1.55,0.5))
```

```python
plt.figure(figsize = (10,4))
sns.swarmplot(data =df,x='math score',y='gender',hue = 'test preparation course',dodge = True)
plt.legend(loc = (1.05,0.5))
```

```python
sns.boxenplot(x='math score',y = 'test preparation course',data = df,hue = 'gender')
```

```python
plt.figure(figsize = (10,10))
sns.violinplot(data =df,x='reading score',y = 'parental level of education',hue = 'test preparation course',inner = 'quartile')
plt.legend(loc = (1.55,0.5))
```
