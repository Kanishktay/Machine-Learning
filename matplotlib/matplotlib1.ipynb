```python
a = np.linspace(0,10,20)
```

```python
b = a**4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y = x*2
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)

axes.set_xlim(0,8)
axes.set_ylim(0,8000)
axes.set_xlabel('a')
axes.set_xlabel('b')
axes.set_title('big')

axess = fig.add_axes([1,0,1,1])
axess.plot(x,y)


axess.set_xlim(0,30)
axess.set_ylim(0,60)
axess.set_xlabel('c')
axess.set_xlabel('d')
axess.set_title('small')
plt.show()
```
