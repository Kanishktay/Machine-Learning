```python
# CODE HERE
def cohort (tenure):
    if tenure<13:
        return '0-12 Months'
    elif tenure<25:
        return '12-24 Months'
    elif tenure<49:
        return '24-48 Months'
    else:
        return 'Ove 48 Months'
```

```python
df['Tenure Cohort'] = df['tenure'].apply(cohort)
```

```python
#CODE HERE
X = df.drop(['Churn' , 'customerID'],axis =1)
X = pd.get_dummies(X,drop_first= True)
```

```python
y = df['Churn']
```

```python
#CODE HERE
from sklearn.model_selection import train_test_split
X_train,X_test, y_train ,y_test  = train_test_split(X,y,test_size=.1 , random_state= 101)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth= 6 )
dt.fit(X_train,y_train)
pred  = dt.predict(X_test)
from sklearn.metrics import classification_report , confusion_matrix, plot_confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
imp_feat = pd.DataFrame(data = dt.feature_importances_ , 
                        index = X.columns , 
                        columns= ['Feat Imp'])
imp_feat  = imp_feat.sort_values('Feat Imp')
imp_feat  = imp_feat[imp_feat['Feat Imp']> 0]
imp_feat
sns.barplot(data =imp_feat , x = imp_feat.index , y = 'Feat Imp')
plt.xticks(rotation = 90)
```

```python
from sklearn.tree import plot_tree
plot_tree(dt)
```

```python
#CODE HERE
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train , y_train)
pred = rf.predict(X_test)
print(classification_report(y_test,pred))
```

```python
plot_confusion_matrix(rf,X_test, y_test)
```

```python
#CODE HERE
from sklearn.ensemble import GradientBoostingClassifier , AdaBoostClassifier
ada_model = AdaBoostClassifier()
ada_model.fit(X_train , y_train)
ada_pred = ada_model.predict(X_test)
gb_model = GradientBoostingClassifier()
gb_model.fit(X_train , y_train)
gb_pred = gb_model.predict(X_test)

print(classification_report(ada_pred , y_test))
print(classification_report(gb_pred , y_test))
```

```python
plot_confusion_matrix(gb_model , X_test, y_test)
plot_confusion_matrix(ada_model , X_test, y_test)
```
